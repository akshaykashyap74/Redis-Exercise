```python
import redis
```

```python
r = redis.Redis(host='localhost', port=6379, decode_responses=True)
print(r)
```

```python
actors = r.keys('actor:*')
movies = r.keys('movie:*')
```

```python
print(f"Number of actors: {len(actors)}")
print(f"Number of movies: {len(movies)}")
```

```python
actors = r.scan_iter("actor:*")
for actor in actors:
    actor_info = r.hgetall(actor)
    if int(actor_info.get('date_of_birth', 0)) < 1980:
        print(actor_info)
```

```python
for movie in movies:
    movie_info = r.hgetall(movie)
    if movie_info['title'] == "The Imitation Game":
        print(f"Genre: {movie_info['genre']}, Rating: {movie_info['rating']}")
```

```python
movie_ratings = []
for movie in movies:
    movie_info = r.hgetall(movie)
    movie_ratings.append((movie_info['title'], float(movie_info['rating'])))
movie_ratings.sort(key=lambda x: x[1], reverse=True)
print(f"Top 5 highest-rated movies: {movie_ratings[:5]}")
```

```python
count = sum(1 for movie in movie_ratings if movie[1] > 7.5)
print(f"Movies with rating above 7.5: {count}")
```

```python
for movie in movies:
    movie_info = r.hgetall(movie)
    if movie_info['title'] == "The Imitation Game":
        r.hset(movie, "rating", 8.5)
```
